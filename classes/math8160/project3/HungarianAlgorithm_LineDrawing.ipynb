```python
import heapq

A= [[0]*4]*4
A[0]=[23,9,0,0]
A[1]=[1,3,2,0]
A[2]=[8,5,4,0]
A[3]=[0,0,0,2]

#initialize list of locations of zeros in A
zeros=[]
for i in range(len(A)):
    for j in range(len(A)):
        if A[i][j]==0:
            zeros.append((i,j))
            
#find the number of zeros in each row and column
row_zeros = []
col_zeros = []
for (r,c) in zeros:
    row_counter = 0
    col_counter = 0
    for (i,j) in zeros:
        if r == i:
            row_counter = row_counter + 1
        if c == j:
            col_counter = col_counter + 1
    row_zeros.append((row_counter,r))
    col_zeros.append((col_counter,c))


# remove duplicates in sets of zeros 
row_zeros=list((set(row_zeros)))
col_zeros=list(set(col_zeros))

#sort the sets from least to greatest
row_zeros.sort()
col_zeros.sort()

print(row_zeros)
print(col_zeros)
print('the column with the most zeros is column ' + str(col_zeros[len(col_zeros)-1][1]))
print('the row with the most zeros is row ' + str(row_zeros[len(row_zeros)-1][1]))

#initialize arrays to store which rows and columns we are drawing lines through
row_lines=[]
col_lines=[]

#draw a line
#check if there is a row with only one zero
if row_zeros[0][0] == 1:
    #find the location of the zero
    for (r,c) in zeros:
        if r==row_zeros[0][1]:
            #draw line through column
            col_lines.append(c)
    #update zeros lists

#check if there is a column with only one zero
elif col_zeros[0][0] == 1:
    #find the location of the zero
    for (r,c) in zeros:
        if c==col_zeros[0][1]:
            #draw line through row
            row_lines.append(r)
    #update zeros lists
    

#if all rows and columns have more than one zero
else:
    #determine if a row or column has the most zeros
    if col_zeros[len(col_zeros)-1][0] > row_zeros[len(row_zeros)-1][0]:
        print('the most zeros are found in column' +str(col_zeros[len(col_zeros)-1][1]))
        #draw line through column
        col_lines.append(col_zeros[len(col_zeros)-1][1])
        #update zeros lists
    else:
        print('the most zeros are found in row' +str(row_zeros[len(row_zeros)-1][1]))
        #draw line through row
        row_lines.append(row_zeros[len(row_zeros)-1][1])
        #update zeros lists
```
