```python
class myItem:
    __slots__ = '_key','_value'

    def __init__(self, key, value):
        """initialize items in pq sorted by key associated with value"""
        self._key = key
        self._value = value

    def __lt__(self, other):
        """overload __lt__ to compare keys"""
        if self._key == other._key:
            return hash(self) < hash(other)
        else:
            return self._key < other._key

    def key(self):
        return self._key

    def value(self):
        return self._value

e1_item=myItem(e1.element(), e1)
e2_item=myItem(e2.element(), e2)
e3_item=myItem(e3.element(), e3)
edgeItems1=[e2_item, e3_item, e1_item]
print(edgeItems1)
heapq.heapify(edgeItems1)
print(e2_item<e3_item)
print(edgeItems1)
for edgeItem in edgeItems1:
    print(edgeItem._key, edgeItem._value, hash(edgeItem))
heapq.heappop(edgeItems1)
print(edgeItems1)
```
